```python
import pytesseract ## 테서랙트 패키지 (이미지 텍스트 추출)
from PIL import Image ## 이미지 관리
from selenium import webdriver ## 크롬 웹드라이버
import time ## 딜레이를 주기 위한 타임 패키지
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pymysql ## mysql 사용
from urllib.request import urlopen ## 이미지 파일 불러오고 저장관련
import os ## 위와 동일
import warnings ## 에러 메시지 무시 함수

def checker(textg): ## 키워드 검출 함수
    if ("국가장학금" or "고지 감면" or "등록금 전액") in textg:
        checknum1 = True
    else:
        checknum1 = False
    print("1번 유형",checknum1)
    if ("생활비성" or "비등록금성" or "학업장려금" or "생활장학금" or "마일리지" or "근로") in textg:
        checknum2 = True
    else:
        checknum2 = False
    print("2번 유형", checknum2)
    if ("학자금 대출" or "대출" or "이자지원" or "상환") in textg:
        checknum3 = True
    else:
        checknum3 = False
    print("3번 유형", checknum3)
    if ("취업연계" or "희망사다리" or "1유형" or "2유형") in textg:
        checknum4 = True
    else:
        checknum4 = False
    print("4번 유형", checknum4)
    return checknum1, checknum2, checknum3, checknum4

def tesserimg(i): ##예외처리로 이미지가 없어 오류가 나도 무시하고 계속 실행 오류 없으면 이미지 있다 판단하여 이미지에서 추출한 택스트 합침
    try:
        ulink = driver.find_element(By.XPATH,'//*[@id="CP1_divContents"]/div/p/img')
        savep = "C:/Users/Ssw11/AppData/Local/Programs/Python/Python312"
        pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'
        src = ulink.get_attribute('src')
        t = urlopen(src).read()
        inum = str(i)
        file = open(os.path.join(savep, inum + '.jpg'),"wb")
        file.write(t)
        imgtext = pytesseract.image_to_string(Image.open('C:/Users/Ssw11/AppData/Local/Programs\Python/Python312/' + inum +".jpg"),lang='kor+eng')
        return imgtext 
    except:
        warnings.filterwarnings(action='ignore')
        return ""

driver = webdriver.Chrome() ## 구글 드라이버 킴
driver.implicitly_wait(3)
driver.get("https://www.dongseo.ac.kr/kr/")

driver.find_element(By.XPATH, '//*[@id="quick-user"]/div/div[1]/h4/a/i[2]').click()
driver.find_element(By.XPATH, '//*[@id="user1"]/div/div[2]/ul/li[1]/a').click()
driver.switch_to.window(driver.window_handles[1])  

## 학교 로그인
driver.find_element(By.NAME, 'id').send_keys('') ##DAP 아이디 입력
driver.find_element(By.NAME, 'pw').send_keys('') ##DAP 비번 입력
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="BtnLogIn"]').click() ##로그인 클릭

## 장학 공지 누르기
driver.find_element(By.XPATH, '//*[@id="Mcont02"]/div[2]/div[1]/div[2]/div[1]/ul/li[1]/a[2]').click()
driver.implicitly_wait(3)
driver.find_element(By.XPATH,'//*[@id="content"]/div[3]/div/ul/li[3]/a').click()
driver.implicitly_wait(10)

## my sql 실행
conn = pymysql.connect(host = 'localhost',
                       user = 'root',
                       password = '', ## 비밀번호 입력
                       db = '', ## db 이름 입력 schema
                       charset = 'utf8')
cursor = conn.cursor()
sql = "INSERT INTO suho (title, link, check1, check2, check3, check4) VALUES (%s, %s, %s, %s, %s, %s)"

for c in range(1,3):
    for j in range(2,5):
        for i in range (1,11):
            if c==2:
                driver.find_element(By.XPATH,'//*[@id="CP1_COM_Page_Controllor1_lbtnNext10"]').click()
                time.sleep(15)
            if j==3:
                pagenum = '//*[@id="CP1_COM_Page_Controllor1_lbtnPage{}"]'.format(j-1)
                nextpage = driver.find_element(By.XPATH,pagenum).click()
                time.sleep(15)
            elif j==4:
                pagenum = '//*[@id="CP1_COM_Page_Controllor1_lbtnPage{}"]'.format(j-1)
                nextpage = driver.find_element(By.XPATH,pagenum).click()
                time.sleep(15)
            peed = '//*[@id="CP1_pnl1"]/div/table/tbody/tr[{}]/td[2]/a'.format(i)
            title = driver.find_element(By.XPATH,peed).text
            link = driver.find_element(By.XPATH,peed).get_attribute("href")
            driver.find_element(By.XPATH,peed).click()
            time.sleep(2)
            textg = driver.find_element(By.XPATH,'//*[@id="CP1_divContents"]/div').text
            imgtxt = tesserimg((c*100 +j*10+i)) ## tesserimg 함수 호출
            textg = imgtxt + textg
            print(textg)
            q,w,e,r = checker(textg) ## 채커 함수 호출
            time.sleep(3)
            cursor.execute(sql, (title,link,q,w,e,r))
            driver.find_element(By.XPATH,'//*[@id="content"]/div[3]/div/div/div/ul/li[3]/a').click()
            time.sleep(15)
    for a in range (1,11):
        if c==2:
            driver.find_element(By.XPATH,'//*[@id="CP1_COM_Page_Controllor1_lbtnNext10"]').click()
            time.sleep(15)
        pagenum = '//*[@id="CP1_COM_Page_Controllor1_lbtnPage{}"]'.format(j)
        nextpage = driver.find_element(By.XPATH,pagenum).click()
        time.sleep(3) ## 4페이지로 넘김
        peed = '//*[@id="CP1_pnl1"]/div/table/tbody/tr[{}]/td[2]/a'.format(a)
        title = driver.find_element(By.XPATH,peed).text
        link = driver.find_element(By.XPATH,peed).get_attribute("href")
        driver.find_element(By.XPATH,peed).click()
        time.sleep(3)
        textg = driver.find_element(By.XPATH,'//*[@id="CP1_divContents"]/div').text
        imgtxt = tesserimg((c* 100 + j*10 + a + 10)) ## tesserimg 함수 호출
        textg = imgtxt + textg
        print(textg)
        q,w,e,r = checker(textg) ## 채커 함수 호출
        time.sleep(3)
        cursor.execute(sql, (title,link,q,w,e,r))
        driver.find_element(By.XPATH,'//*[@id="content"]/div[3]/div/div/div/ul/li[3]/a').click()
        time.sleep(15)
    for b in range (1,11):
        if c==2:
            driver.find_element(By.XPATH,'//*[@id="CP1_COM_Page_Controllor1_lbtnNext10"]').click()
            time.sleep(15)
        pagenum = '//*[@id="CP1_COM_Page_Controllor1_lbtnPage{}"]'.format(10) ## 5페이지 로 넘김
        nextpage = driver.find_element(By.XPATH,pagenum).click()
        time.sleep(3)
        peed = '//*[@id="CP1_pnl1"]/div/table/tbody/tr[{}]/td[2]/a'.format(b)
        title = driver.find_element(By.XPATH,peed).text
        link = driver.find_element(By.XPATH,peed).get_attribute("href")
        driver.find_element(By.XPATH,peed).click()
        time.sleep(3)
        textg = driver.find_element(By.XPATH,'//*[@id="CP1_divContents"]/div').text
        imgtxt = tesserimg((c*100+j*10+b +20)) ## tesserimg 함수 호출
        textg = imgtxt + textg
        print(textg)
        q,w,e,r = checker(textg) ## 채커 함수 호출
        time.sleep(3)
        cursor.execute(sql, (title,link,q,w,e,r))
        driver.find_element(By.XPATH,'//*[@id="content"]/div[3]/div/div/div/ul/li[3]/a').click()
        time.sleep(15)
    
conn.commit()
conn.close()
driver.close()
```
